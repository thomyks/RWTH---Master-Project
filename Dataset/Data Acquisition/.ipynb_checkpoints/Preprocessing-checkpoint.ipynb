```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# replace file path with custom one if necessary
df = pd.read_csv('complete_dataset_uncleaned.csv')
```

```python
# new column 'entities' generated by splitting entity_strings column of type str into list
df['entities'] = df.entity_strings.str.split(',')
# new column 'nr_entities' with count of entities
df['nr_entities'] = df['entities'].str.len()
df.fillna({'nr_entities': 0}, inplace=True)
```

```python
df.sort_values(by='nr_entities', inplace=True)
print(len(df))
```

```python
# drop rows where the claim url is duplicate
df.drop_duplicates(subset='claim', keep='last', inplace=True)
print(len(df))
```

```python
# drop rows where the claim text is duplicate
df.drop_duplicates(subset='text', keep='last', inplace=True)
print(len(df))
```

```python
# drop claims which are shorter than two words (just split by whitespace)
df['num_words'] = df.text.map(lambda text: len(text.split(' ')))
```

```python
df = df[df.num_words > 2]
```

```python
# install langdetect library if necessary
# !pip install langdetect
```

```python
from langdetect import detect
```

```python
df['probable_lang'] = df.text.map(lambda text: detect(text))
```

```python
# drop non english claims
english_df = df[df.probable_lang == 'en']
```

```python
print(len(english_df))
```

```python
# save new df
english_df.to_csv('preprocessed_df.csv')
```

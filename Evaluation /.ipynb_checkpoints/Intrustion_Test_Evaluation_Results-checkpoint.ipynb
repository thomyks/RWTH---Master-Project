```python
import pandas as pd
import json
```

```python
# specify filepaths of two different testControl jsons files
fp1 = ''
fp2 = ''
```

```python
with open(fp1) as fp:
    results1 = json.load(fp)

with open(fp2) as fp:
    results2 = json.load(fp)
```

```python
set_names = [result['test_set_name'] for result in results1['results']]
correct = [result['actual_intruder_index'] for result in results1['results']]
rater1 = [result['selected_intruder_index'] for result in results1['results']]
rater2 = [result['selected_intruder_index'] for result in results2['results']]
```

```python
df = pd.DataFrame(data = {'set_name': set_names, 'correct': correct, 'rater1': rater1, 'rater2': rater2})
```

```python
df = df.astype({'rater1': 'int', 'rater2': 'int'})
```

```python
df.head(28)
```

```python
rater1_correct = sum(df.correct == df.rater1)
rater2_correct = sum(df.correct == df.rater2)

perc_rater1_correct = rater1_correct/len(df)
perc_rater2_correct = rater2_correct/len(df)

raters_agreeing = sum(df.rater1 == df.rater2)
perc_raters_agreeing = raters_agreeing/len(df)

raters_both_correct = sum(df.apply(lambda row: row.correct == row.rater1 == row.rater2, axis = 1))
perc_raters_both_correct = raters_both_correct/len(df)

raters_consistent_incorrect = sum(df.apply(lambda row: row.correct != row.rater1 == row.rater2, axis = 1))
perc_raters_consistent_incorrect = raters_consistent_incorrect/len(df)

raters_only_one_correct = sum(df.apply(lambda row: (row.correct == row.rater1 and row.correct != row.rater2 and row.rater1 != row.rater2) or 
                   (row.correct != row.rater1 and row.correct == row.rater2 and row.rater1 != row.rater2) , axis = 1))
perc_raters_only_one_correct = raters_only_one_correct/len(df)

raters_consistent_incorrect = sum(df.apply(lambda row: (row.correct != row.rater1) and (row.rater1 == row.rater2), axis = 1))
perc_raters_consistent_incorrect = raters_consistent_incorrect/len(df)

raters_inconsistent_incorrect = sum(df.apply(lambda row: (row.correct != row.rater1) and (row.correct != row.rater2) and (row.rater1 != row.rater2), axis = 1))
perc_raters_inconsistent_incorrect = raters_inconsistent_incorrect/len(df)


print('percentage rater1 correct')
print(perc_rater1_correct)

print('percentage rater2 correct')
print(perc_rater2_correct)

print('percentage raters agreeing')
print(perc_raters_agreeing)

print('percentage raters both correct')
print(perc_raters_both_correct)

print('percentage raters only one correct')
print(perc_raters_only_one_correct)

print('percentage raters consistent incorrect')
print(perc_raters_consistent_incorrect)

print('percentage raters inconsistent incorrect')
print(perc_raters_inconsistent_incorrect)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(facecolor='white')

# might be neccessary to only plot non zero parts of pie chart
plt.pie([perc_raters_both_correct, perc_raters_only_one_correct, perc_raters_inconsistent_incorrect, perc_raters_consistent_incorrect],
       labels = ['both correct', 'one correct', 'both incorrect, different', 'both incorrect, same'],
    autopct = lambda perc: round(perc, 1),
       startangle = 90,
       radius = 1.5,
       textprops = {'size': 12},
        colors=['#007cac', 'C1', 'C2', 'C3']
       )

plt.title('Intrusion Test Results', pad = 60, fontdict= {'size': 20})



plt.show()
```
